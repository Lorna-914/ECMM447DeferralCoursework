```python
# Load spaCy model
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# by deepseek
# Named Entity Recognition
def extract_entities(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    print("Named Entities:", entities)
    return entities # Return the entities list
```

```python
# by deepseek
input_file = "/content/3_The Beauty Myth.txt"

text  = open(input_file, 'r', encoding='utf-8').read()
doc = nlp(text)
sentences = [sent.text for sent in doc.sents]
entities = extract_entities(text) # Assign the returned entities
```

```python
# by Gemini
from collections import defaultdict
import pandas as pd

# Group entities by label
grouped_entities = defaultdict(list)
for entity, label in entities:
    grouped_entities[label].append(entity)

# Convert to a pandas DataFrame for better display
df_entities = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in grouped_entities.items()]))

# Display the DataFrame
display(df_entities)
```
